```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
quality = ctrl.Antecedent(np.arange(0,11,1), 'quality')
service = ctrl.Antecedent(np.arange(0,11,1), 'service')
tip=ctrl.Consequent(np.arange(0,26,1), 'tip')
quality.automf(3)
service.automf(3)
tip['very low'] = fuzz.trimf(tip.universe, [0, 0, 5])
tip['low']= fuzz.trimf(tip.universe,[0,5,13])
tip['medium']= fuzz.trimf(tip.universe,[0,13,25])
tip['high']= fuzz.trimf(tip.universe,[13,25,25])
tip['very high'] = fuzz.trimf(tip.universe, [20, 25, 25])
```

```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
marks = np.arange(0, 101, 1)
grades = np.arange(0, 11, 1)
# Define fuzzy membership functions for marks
poor = fuzz.trimf(marks, [0, 0, 50])
average = fuzz.trimf(marks, [40, 50, 60])
good = fuzz.trimf(marks, [50, 75, 100])
excellent = fuzz.trimf(marks, [75, 100, 100])
# Define fuzzy membership functions for grades
fail = fuzz.trimf(grades, [0, 0, 4])
pass_ = fuzz.trimf(grades, [3, 5, 7])
merit = fuzz.trimf(grades, [6, 8, 10])
distinction = fuzz.trimf(grades, [8, 10, 10])
# Plot the fuzzy membership functions
fig, (ax0, ax1) = plt.subplots(nrows=2, figsize=(8, 10))
ax0.plot(marks, poor, 'r', linewidth=1.5, label='Poor')
ax0.plot(marks, average, 'b', linewidth=1.5, label='Average')
ax0.plot(marks, good, 'g', linewidth=1.5, label='Good')
ax0.plot(marks, excellent, 'purple', linewidth=1.5, label='Excellent')
ax0.set_title('Marks Fuzzy Membership Functions')
ax0.legend()
ax1.plot(grades, fail, 'r', linewidth=1.5, label='Fail')
ax1.plot(grades, pass_, 'b', linewidth=1.5, label='Pass')
ax1.plot(grades, merit, 'g', linewidth=1.5, label='Merit')
ax1.plot(grades, distinction, 'purple', linewidth=1.5, label='Distinction')
ax1.set_title('Grades Fuzzy Membership Functions')
ax1.legend()
# Display the plots
plt.tight_layout()
plt.show()
```
